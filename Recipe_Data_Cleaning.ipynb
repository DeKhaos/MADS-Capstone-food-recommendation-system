```python
def dataset_summary(df):
    
    def safe_nunique(col):
        try:
            return col.nunique()
        except TypeError:
            return col.astype(str).nunique()
    
    summary = pd.DataFrame({
        "dtype": df.dtypes,
        "missing_count": df.isnull().sum(),
        "missing_pct": (df.isnull().sum() / len(df) * 100).round(2),
        "n_unique": df.apply(safe_nunique)
    }).sort_values(by="missing_count", ascending=False)

    return summary
```

```python
# fill in missing value

def fill_na_with_value(df, columns, value=0):
    
    for col in columns:
        if col in df.columns:
            df[col] = df[col].fillna(value)
    
    return df
```

```python
#reformat columns with c() to list of strings
def parse_c_vector(x):
    
    if pd.isna(x):
        return []
    
    x = str(x).strip()
    
    if x.startswith("c(") and x.endswith(")"):
        x = x[2:-1]   # remove c( )
        items = re.findall(r'"(.*?)"', x)
        return items
    
    return []

def clean_c_columns(df, columns):

    for col in columns:
        if col in df.columns:
            df[col] = df[col].apply(parse_c_vector)

    return df
```

```python
#convert prepTime and cookTime to minutes
def iso_to_minutes(x):
    if pd.isna(x):
        return np.nan
    
    x = str(x)
    h = re.search(r"(\d+)H", x)
    m = re.search(r"(\d+)M", x)

    hours = int(h.group(1)) if h else 0
    mins = int(m.group(1)) if m else 0

    return hours * 60 + mins

def add_time_minutes(df, columns):
    for col in columns:
        if col in df.columns:
            df[col + "_Minutes"] = df[col].apply(iso_to_minutes)
    return df

def clean_date_column(df, column):
    if column in df.columns:
        df[column + "_Cleaned"] = pd.to_datetime(df[column], errors="coerce")
    return df
```

```python
import pandas as pd
import numpy as np
import re
from difflib import SequenceMatcher

DESCRIPTOR_WORDS = {
    "fresh", "dried", "organic", "raw", "ripe", "pure", "natural",
    "chopped", "diced", "sliced", "minced", "crushed", "shredded",
    "grated", "ground", "peeled", "trimmed", "cubed", "halved",
    "julienned", "thinly", "finely", "roughly",
    "cooked", "uncooked", "boiled", "fried", "roasted", "toasted",
    "baked", "steamed", "grilled",
    "soft", "firm", "smooth", "thick", "thin",
    "boneless", "skinless", "lean", "extra", "virgin",
    "large", "small", "medium", "baby",
    "optional", "divided", "packed", "light", "dark", "plain"
}

PROTECTED_INGREDIENTS = {
    "brown sugar",
    "green onion",
    "coconut milk",
    "soy sauce",
    "sesame oil",
    "olive oil",
    "vegetable oil",
    "chili pepper",
    "green chili pepper",
    "bell pepper",
    "green bell pepper",
    "red bell pepper",
    "cream cheese",
    "peanut butter",
    "sweet potato"
}

def explode_ingredients(df, ingredient_col="RecipeIngredientParts"):
    rows = []

    for recipe_id, ingredients in zip(df["RecipeId"], df[ingredient_col]):
        if isinstance(ingredients, list):
            for ing in ingredients:
                if isinstance(ing, str) and ing.strip():
                    rows.append({
                        "RecipeId": recipe_id,
                        "raw_ingredient": ing.strip()
                    })

    return pd.DataFrame(rows)

def normalize_text(text):
    if not isinstance(text, str):
        return ""

    text = text.lower().strip()
    text = re.sub(r"[^a-z\s]", " ", text)

    for word in DESCRIPTOR_WORDS:
        text = re.sub(rf"\b{re.escape(word)}\b", "", text)

    text = re.sub(r"\s+", " ", text).strip()
    return text

def simple_singularize(text):
    if not isinstance(text, str) or not text:
        return ""

    words = text.split()
    if not words:
        return text

    last = words[-1]

    if last.endswith("ies") and len(last) > 3:
        last = last[:-3] + "y"
    elif last.endswith("oes") and len(last) > 3:
        last = last[:-2]
    elif last.endswith("s") and not last.endswith("ss") and len(last) > 3:
        last = last[:-1]

    words[-1] = last
    return " ".join(words)

def normalize_ingredient_phrase(text):
    text = normalize_text(text)
    text = simple_singularize(text)
    return text

def ingredient_frequency_table(ingredient_df, col="normalized"):
    freq = ingredient_df[col].value_counts().reset_index()
    freq.columns = [col, "count"]
    return freq

def get_head_word(text):
    if not isinstance(text, str) or not text.strip():
        return ""
    return text.split()[-1]

def string_similarity(a, b):
    return SequenceMatcher(None, a, b).ratio()

def build_candidate_synonym_pairs(freq_table, min_count=20, min_similarity=0.80):
    freq_dict = dict(zip(freq_table["normalized"], freq_table["count"]))
    ingredients = [x for x in freq_table["normalized"] if freq_dict[x] >= min_count]

    candidates = []

    for i in range(len(ingredients)):
        for j in range(i + 1, len(ingredients)):
            a = ingredients[i]
            b = ingredients[j]

            same_head = get_head_word(a) == get_head_word(b)
            sim = string_similarity(a, b)

            if same_head or sim >= min_similarity:
                candidates.append({
                    "ingredient_a": a,
                    "ingredient_b": b,
                    "count_a": freq_dict[a],
                    "count_b": freq_dict[b],
                    "same_head": same_head,
                    "similarity": round(sim, 3)
                })

    candidates_df = pd.DataFrame(candidates)

    if not candidates_df.empty:
        candidates_df = candidates_df.sort_values(
            by=["same_head", "similarity", "count_a", "count_b"],
            ascending=[False, False, False, False]
        )

    return candidates_df

def choose_canonical_by_frequency(group, freq_dict):
    group = list(group)
    return max(group, key=lambda x: freq_dict.get(x, 0))

def build_headword_synonym_map(freq_table, min_count=20, max_group_size=10):
    freq_dict = dict(zip(freq_table["normalized"], freq_table["count"]))

    grouped = {}
    for ing in freq_table["normalized"]:
        if freq_dict[ing] >= min_count:
            head = get_head_word(ing)
            grouped.setdefault(head, []).append(ing)

    synonym_map = {}

    for head, items in grouped.items():
        items = sorted(items, key=lambda x: freq_dict[x], reverse=True)

        if 1 < len(items) <= max_group_size:
            canonical = choose_canonical_by_frequency(items, freq_dict)

            for item in items:
                if item != canonical:
                    synonym_map[item] = canonical

    return synonym_map

def filter_synonym_map(synonym_map, protected_set):
    filtered = {}

    for k, v in synonym_map.items():
        if k in protected_set or v in protected_set:
            continue
        filtered[k] = v

    return filtered

def apply_synonym_map(ingredients, synonym_map):
    if not isinstance(ingredients, list):
        return []

    mapped = []
    for ing in ingredients:
        canon = synonym_map.get(ing, ing)
        if canon:
            mapped.append(canon)

    return list(dict.fromkeys(mapped))
```

```python
ingredient_df = explode_ingredients(recipes_v3, "RecipeIngredientParts")
ingredient_df["normalized"] = ingredient_df["raw_ingredient"].apply(normalize_ingredient_phrase)

freq_table = ingredient_frequency_table(ingredient_df, "normalized")

candidate_pairs = build_candidate_synonym_pairs(
    freq_table,
    min_count=30,
    min_similarity=0.80
)

auto_synonyms = build_headword_synonym_map(
    freq_table,
    min_count=30,
    max_group_size=10
)

auto_synonyms_filtered = filter_synonym_map(
    auto_synonyms,
    PROTECTED_INGREDIENTS
)

recipes_v4 = recipes_v3.copy()
recipes_v4["ingredients_canonical_auto"] = recipes_v4["ingredients_canonical"].apply(
    lambda x: apply_synonym_map(x, auto_synonyms_filtered)
)
```

```python



combined_ingredient_map = ingredient_map.copy()
combined_ingredient_map.update(APPROVED_INGREDIENT_MAP)


# Update with the stronger canonical map
combined_ingredient_map.update(CANONICAL_INGREDIENT_MAP)

FINAL_FIXES = {
    "lemons rind of": "lemon zest",
    "lemon zest of": "lemon zest",
    "rind of lemon": "lemon zest",
    "zest of lemon": "lemon zest",
    "lime rind": "lime zest",
    "rind of lime": "lime zest",
    "zest of lime": "lime zest",
    "orange rind": "orange zest",
    "rind of orange": "orange zest",
    "zest of orange": "orange zest",
    "mint leaf": "mint",
}
def apply_combined_ingredient_map(ingredients, ingredient_map, final_fixes=None):
    if not isinstance(ingredients, list):
        return []
    
    mapped = [ingredient_map.get(i, i) for i in ingredients]
    
    if final_fixes is not None:
        mapped = [final_fixes.get(i, i) for i in mapped]
    
    # remove duplicates while keeping order
    return list(dict.fromkeys(mapped))

recipes_v4["ingredients_canonical_final"] = recipes_v4["ingredients_canonical"].apply(
    lambda x: apply_combined_ingredient_map(
        x,
        combined_ingredient_map,
        final_fixes=FINAL_FIXES
    )
)
```

```python
def summarize_ingredients_column(df, col="ingredients_canonical_final"):
    
    total_recipes = len(df)
    
    ingredient_lists = df[col].dropna()
    
    total_ingredient_mentions = sum(len(i) for i in ingredient_lists if isinstance(i, list))
    
    all_ingredients = [
        ing
        for lst in ingredient_lists
        if isinstance(lst, list)
        for ing in lst
    ]
    
    unique_ingredients = set(all_ingredients)
    
    print("DATASET SUMMARY")
    print("------------------------")
    print("Total recipes:", total_recipes)
    print("Recipes with ingredient lists:", len(ingredient_lists))
    print("Total ingredient mentions:", total_ingredient_mentions)
    print("Unique ingredients:", len(unique_ingredients))
    print("Average ingredients per recipe:",
          round(total_ingredient_mentions / len(ingredient_lists), 2))
    
    return all_ingredients
```

```python
all_ingredients = summarize_ingredients_column(recipes_v4)

ingredient_exploded = recipes_v4[["RecipeId", "ingredients_canonical_final"]].explode(
    "ingredients_canonical_final"
)

ingredient_summary = (
    ingredient_exploded
    .groupby("ingredients_canonical_final")
    .agg(
        count=("RecipeId", "count"),
        #recipe_ids=("RecipeId", lambda x: list(x))
    )
    .reset_index()
    .sort_values("count", ascending=False)
)
```
